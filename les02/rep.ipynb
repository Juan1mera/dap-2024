````python
class Solution1:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        seen = {}
        for i, num in enumerate(nums):
            complement = target - num
            if complement in seen:
                return [seen[complement], i]
            seen[num] = i
        return []

# Example usage
solution = Solution1()

# Test case 1
nums1 = [2, 7, 11, 15]
target1 = 9
result1 = solution.twoSum(nums1, target1)
print(f"Test 1: nums = {nums1}, target = {target1}, output = {result1}")

# Test case 2
nums2 = [3, 2, 4]
target2 = 6
result2 = solution.twoSum(nums2, target2)
print(f"Test 2: nums = {nums2}, target = {target2}, output = {result2}")

# Test case 3
nums3 = [3, 3]
target3 = 6
result3 = solution.twoSum(nums3, target3)
print(f"Test 3: nums = {nums3}, target = {target3}, output = {result3}")

````

````python
class Solution2(object):
    def isPalindrome(self, x):
        """
        :type x: int
        :rtype: bool
        """
        if x < 0:
            return False

        digits = []
        while x > 0:
            digits.append(x % 10)
            x //= 10

        left, right = 0, len(digits) - 1
        while left < right:
            if digits[left] != digits[right]:
                return False
            left += 1
            right -= 1

        return True


solution = Solution2()

# Test case 1
x1 = 121
result1 = solution.isPalindrome(x1)
print(f"Test 1: x = {x1}, output = {result1}")

# Test case 2
x2 = -121
result2 = solution.isPalindrome(x2)
print(f"Test 2: x = {x2}, output = {result2}")

# Test case 3
x3 = 10
result3 = solution.isPalindrome(x3)
print(f"Test 3: x = {x3}, output = {result3}")
````

````python
class Solution3(object):
    def romanToInt(self, s):
        """
        :type s: str
        :rtype: int
        """
        roman_values = {
            'I': 1,
            'V': 5,
            'X': 10,
            'L': 50,
            'C': 100,
            'D': 500,
            'M': 1000
        }

        result = 0
        prev_value = 0

        for char in reversed(s):
            curr_value = roman_values[char]
            if curr_value >= prev_value:
                result += curr_value
            else:
                result -= curr_value
            prev_value = curr_value

        return result


solution = Solution3()

# Test case 1
s1 = "III"
result1 = solution.romanToInt(s1)
print(f"Test 1: s = {s1}, output = {result1}")

# Test case 2
s2 = "LVIII"
result2 = solution.romanToInt(s2)
print(f"Test 2: s = {s2}, output = {result2}")

# Test case 3
s3 = "MCMXCIV"
result3 = solution.romanToInt(s3)
print(f"Test 3: s = {s3}, output = {result3}")
````

````python
class Solution4(object):
    def longestCommonPrefix(self, strs):
        """
        :type strs: List[str]
        :rtype: str
        """
        if not strs:
            return ""

        # Take the first string as reference
        prefix = strs[0]

        # Compare with other strings
        for string in strs[1:]:
            # Reduce prefix length while it doesn't match the start of current string
            while string[:len(prefix)] != prefix:
                prefix = prefix[:-1]
                if not prefix:
                    return ""

        return prefix


solution = Solution4()

# Test case 1
strs1 = ["flower", "flow", "flight"]
result1 = solution.longestCommonPrefix(strs1)
print(f"Test 1: strs = {strs1}, output = {result1}")

# Test case 2
strs2 = ["dog", "racecar", "car"]
result2 = solution.longestCommonPrefix(strs2)
print(f"Test 2: strs = {strs2}, output = {result2}")

````

````python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution5(object):
    def mergeTwoLists(self, list1, list2):
        """
        :type list1: Optional[ListNode]
        :type list2: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        # Create a dummy node to serve as the head of the merged list
        dummy = ListNode(0)
        current = dummy

        # While both lists have nodes
        while list1 and list2:
            if list1.val <= list2.val:
                current.next = list1
                list1 = list1.next
            else:
                current.next = list2
                list2 = list2.next
            current = current.next

        # Attach remaining nodes from list1 or list2
        if list1:
            current.next = list1
        if list2:
            current.next = list2

        return dummy.next

# Example usage
def create_linked_list(arr):
    if not arr:
        return None
    head = ListNode(arr[0])
    current = head
    for val in arr[1:]:
        current.next = ListNode(val)
        current = current.next
    return head

def linked_list_to_list(head):
    result = []
    current = head
    while current:
        result.append(current.val)
        current = current.next
    return result


solution = Solution5()

# Test case 1
list1 = create_linked_list([1, 2, 4])
list2 = create_linked_list([1, 3, 4])
result = solution.mergeTwoLists(list1, list2)
print(f"Test 1: list1 = [1,2,4], list2 = [1,3,4], output = {linked_list_to_list(result)}")

# Test case 2
list1 = create_linked_list([])
list2 = create_linked_list([])
result = solution.mergeTwoLists(list1, list2)
print(f"Test 2: list1 = [], list2 = [], output = {linked_list_to_list(result)}")

# Test case 3
list1 = create_linked_list([])
list2 = create_linked_list([0])
result = solution.mergeTwoLists(list1, list2)
print(f"Test 3: list1 = [], list2 = [0], output = {linked_list_to_list(result)}")

````

````python
class Solution6(object):
    def lengthOfLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """
        seen = {}  # Dictionary to store character and its latest index
        max_length = 0
        start = 0

        for end, char in enumerate(s):
            if char in seen and seen[char] >= start:
                # Move start to the position after the last occurrence of char
                start = seen[char] + 1
            else:
                # Update max_length if current window is longer
                max_length = max(max_length, end - start + 1)
            # Update the latest index of the current character
            seen[char] = end

        return max_length


solution = Solution6()

# Test case 1
s1 = "abcabcbb"
result1 = solution.lengthOfLongestSubstring(s1)
print(f"Test 1: s = {s1}, output = {result1}")

# Test case 2
s2 = "bbbbb"
result2 = solution.lengthOfLongestSubstring(s2)
print(f"Test 2: s = {s2}, output = {result2}")

# Test case 3
s3 = "pwwkew"
result3 = solution.lengthOfLongestSubstring(s3)
print(f"Test 3: s = {s3}, output = {result3}")

````

# Ответы на вопросы по программированию на Python

## 1. Что такое динамическая типизация и чем она отличается от статической?

- **Динамическая типизация**: В языках, таких как Python, типы переменных определяются во время выполнения программы. Не нужно явно указывать тип переменной. Например, `x = 5` может стать `x = "текст"` без ошибок.
- **Статическая типизация**: В языках, таких как C++ или Java, типы переменных должны быть явно объявлены (например, `int x = 5`) и не могут изменяться во время выполнения.
- **Отличия**:
  - **Время проверки типов**: Динамическая — во время выполнения, статическая — во время компиляции.
  - **Гибкость**: Динамическая позволяет менять типы, статическая более строгая.
  - **Ошибки**: В статической типизации ошибки типов обнаруживаются до выполнения, в динамической — во время выполнения.

## 2. Чем отличается цикл for в Python от цикла for в C++?

- **Python**:
  - Основан на итерации по последовательностям (списки, кортежи и т.д.) или итераторам: `for x in список`.
  - Не требует явного счетчика, хотя можно использовать `range()`.
  - Прост и читаем, ориентирован на прямой перебор элементов.
  - Пример: `for x in [1, 2, 3]: print(x)` выводит 1, 2, 3.
- **C++**:
  - Более явный, использует счетчик или условие: `for (int i = 0; i < n; i++)`.
  - Требует определения инициализации, условия и обновления счетчика.
  - Более гибкий для управления потоком, но менее интуитивен для перебора коллекций.
  - Пример: `for (int i = 0; i < 3; i++) cout << i;` выводит 0, 1, 2.


## 3. Для чего предназначены операторы continue и break?

- **`continue`**: Пропускает оставшуюся часть текущей итерации цикла и переходит к следующей. Используется, чтобы игнорировать определенные итерации.
  - Пример: `for i in range(5): if i == 2: continue; print(i)` пропустит 2, выведет 0, 1, 3, 4.
- **`break`**: Прерывает цикл полностью, выходя из него. Используется для досрочного завершения цикла.
  - Пример: `for i in range(5): if i == 3: break; print(i)` выведет 0, 1, 2.

## 4. Как определить функцию в Python? Может ли функция возвращать несколько значений?

- **Определение функции**:
  - Используется ключевое слово `def`, за которым следует имя функции, параметры в скобках и тело функции:
    ```python
    def имя_функции(параметр1, параметр2):
        # тело функции
        return результат
    ```
  - Пример: `def add(a, b): return a + b`.
- **Возврат нескольких значений**:
  - Да, функция может возвращать несколько значений, используя кортеж, список или распаковку.
  - Пример:
    ```python
    def get_values():
        return 1, 2, 3  # Возвращает кортеж (1, 2, 3)
    a, b, c = get_values()  # Распаковка результата
    ```

## 5. Что такое лямбда-функции и для чего они предназначены?

- **Лямбда-функции**: Это анонимные функции, определяемые с помощью ключевого слова `lambda`. Они используются для создания кратких функций в одну строку.
  - Синтаксис: `lambda аргументы: выражение`.
  - Пример: `f = lambda x: x * 2` создает функцию, удваивающую `x`.
- **Назначение**:
  - Используются для кратких операций, часто в качестве аргументов для функций высшего порядка, таких как `map()`, `filter()`, или `sorted()`.
  - Пример: `sorted([(1, 'b'), (2, 'a')], key=lambda x: x[1])` сортирует по второму элементу.

## 6. Может ли тело функции содержать определение другой функции?

- **Да**, в Python функции могут быть вложенными (внутренние функции определяются внутри тела другой функции).
  - Пример:
    ```python
    def outer():
        def inner():
            print("Внутренняя функция")
        inner()
    outer()  # Выводит: Внутренняя функция
  ```
- **Назначение**: Вложенные функции полезны для инкапсуляции логики, замыканий (closures) или декораторов.

## 7. Как определить класс в Python? Для чего нужна переменная self?

- **Определение класса**:
  - Используется ключевое слово `class`, за которым следует имя класса и тело:
    ```python
    class MyClass:
        def __init__(self, param):
            self.param = param
        def method(self):
            return self.param
    ```
- **Переменная `self`**:
  - `self` — это ссылка на текущий экземпляр класса. Она передается автоматически при вызове методов экземпляра и используется для доступа к атрибутам и методам объекта.
  - Пример: `obj = MyClass(5); print(obj.method())` использует `self` для доступа к `param`.

## 8. Как определить приватный метод? Можно ли вызвать приватный метод за пределами класса?

- **Определение приватного метода**:
  - Приватные методы обозначаются с помощью двойного подчеркивания `__` перед именем: `__method`.
  - Пример:
    ```python
    class MyClass:
        def __private_method(self):
            print("Это приватный метод")
    ```
- **Вызов за пределами класса**:
  - Технически приватные методы в Python не являются полностью приватными (это соглашение, а не строгая изоляция). Их можно вызвать, используя `_ИмяКласса__имя_метода`, например: `obj._MyClass__private_method()`.
  - Однако это считается плохой практикой, так как нарушает инкапсуляцию.

## 9. Как определить статический метод класса? Для чего может понадобиться статический метод?

- **Определение статического метода**:
  - Используется декоратор `@staticmethod`:
    ```python
    class MyClass:
        @staticmethod
        def static_method():
            print("Статический метод")
    ```
  - Вызывается через класс: `MyClass.static_method()`.
- **Назначение**:
  - Статические методы не требуют доступа к экземпляру (`self`) или классу (`cls`).
  - Используются для утилитарных функций, логически связанных с классом, но не зависящих от его состояния.
  - Пример: метод для проверки формата данных, связанного с классом.

## 10. Чем staticmethod отличается от classmethod?

- **`staticmethod`**:
  - Не принимает ни `self`, ни `cls` в качестве первого аргумента.
  - Не имеет доступа к состоянию класса или экземпляра.
  - Используется для функций, которые логически принадлежат классу, но не зависят от него.
  - Пример: `MyClass.static_method()`.
- **`classmethod`**:
  - Принимает `cls` (класс) в качестве первого аргумента.
  - Может обращаться к атрибутам класса или вызывать другие методы класса.
  - Используется для работы с классом, например, для альтернативных конструкторов.
  - Пример:
    ```python
    class MyClass:
        @classmethod
        def from_string(cls, s):
            return cls(s)
    ```
- **Ключевые отличия**:
  - `staticmethod` — независимая функция в пространстве имен класса.
  - `classmethod` — работает с классом и его атрибутами.

## 11. Как определить собственный класс исключений в Python? Как выбросить и перехватить исключение?

- **Определение класса исключений**:
  - Создается класс, наследующий от `Exception` или другого встроенного исключения:
    ```python
    class MyCustomError(Exception):
        pass
    ```
- **Выброс исключения**:
  - Используется `raise`:
    ```python
    raise MyCustomError("Произошла ошибка")
    ```
- **Перехват исключения**:
  - Используется конструкция `try`/`except`:
    ```python
    try:
        raise MyCustomError("Ошибка")
    except MyCustomError as e:
        print(f"Поймано исключение: {e}")
    ```

## 12. Что такое кортеж, для чего он используется?

- **Кортеж (tuple)**: Это неизменяемая упорядоченная последовательность элементов, определяемая с помощью круглых скобок: `(1, 2, 3)` или `tuple([1, 2, 3])`.
- **Использование**:
  - Хранение фиксированных данных, которые не должны изменяться.
  - Возврат нескольких значений из функции: `return 1, 2`.
  - Использование в качестве ключей словаря (в отличие от списков, кортежи неизменяемы и хэшируемы).
  - Экономия памяти по сравнению со списками.
  - Пример: `t = (1, "text", 3.14); print(t[0])` выводит 1.
