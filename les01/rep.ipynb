## 1. Типы ячеек в Jupyter Notebook

В Jupyter Notebook существует три основных типа ячеек:

- **Code (Код)** - для написания и выполнения программного кода (Python, R, Julia и др.)
- **Markdown (Текст)** - для форматированного текста с использованием Markdown разметки
- **Raw (Сырой текст)** - для текста, который не должен преобразовываться ядром Notebook

## 2. Как с помощью Markdown сформировать иерархический список?

Иерархический список создается с помощью отступов:

```markdown
- Уровень 1
  - Уровень 2
    - Уровень 3
  - Уровень 2 снова
- Возврат к уровню 1
```

Или для нумерованных списков:
```markdown
1. Первый пункт
  1.1. Подпункт
  1.2. Еще подпункт
2. Второй пункт
```

## 3. Как с помощью Markdown оформить цитату?

Цитаты оформляются с помощью символа `>`:

```markdown
> Это цитата первого уровня
>> Это вложенная цитата
> Возврат к первому уровню
```

Результат:
> Это цитата первого уровня
>> Это вложенная цитата
> Возврат к первому уровню

## 4. Как в текстовую ячейку вставить отформатированный блок кода?

Используются тройные обратные апострофы с указанием языка:

````markdown
```python
def hello_world():

    return True

````

Или без указания языка:
````markdown
```

```
````

## 5. Сколько уровней заголовка существует в Markdown?

В Markdown существует 6 уровней заголовков:

```markdown
# Заголовок 1 уровня (H1)
## Заголовок 2 уровня (H2)
### Заголовок 3 уровня (H3)
#### Заголовок 4 уровня (H4)
##### Заголовок 5 уровня (H5)
###### Заголовок 6 уровня (H6)
```

## 6. Шаги процесса анализа данных

Основные шаги процесса анализа данных (CRISP-DM):

1. **Понимание бизнес-задачи** (Business Understanding)
2. **Понимание данных** (Data Understanding)
3. **Подготовка данных** (Data Preparation)
4. **Моделирование** (Modeling)
5. **Оценка** (Evaluation)
6. **Внедрение** (Deployment)

### Подробно о шаге "Подготовка данных"

**Подготовка данных** - наиболее трудоемкий этап, занимающий 60-80% времени проекта. Включает:

- **Очистка данных**: обработка пропущенных значений, выбросов, дубликатов
- **Преобразование данных**: нормализация, стандартизация, кодирование категориальных переменных
- **Инженерия признаков**: создание новых переменных на основе существующих
- **Интеграция данных**: объединение данных из разных источников
- **Снижение размерности**: PCA, feature selection

Качество подготовки данных напрямую влияет на результаты последующего анализа и моделирования.

## 7. Типы данных

Основные типы данных в анализе:

- **Числовые** (Numerical):
  - Дискретные (целые числа)
  - Непрерывные (дробные числа)
- **Категориальные** (Categorical):
  - Номинальные (без порядка)
  - Порядковые (с порядком)
- **Текстовые** (Text)
- **Временные ряды** (Time Series)
- **Геопространственные** (Geospatial)

### Подробно о категориальных данных

**Категориальные данные** представляют характеристики, которые можно разделить на группы. Особенности:

- **Номинальные данные**: категории без естественного порядка (цвета, города, типы продуктов)
- **Порядковые данные**: категории с естественным порядком (уровни образования: школа, бакалавр, магистр)

**Методы обработки**:
- One-Hot Encoding для номинальных данных
- Label Encoding для порядковых данных
- Frequency Encoding заменой на частоту встречаемости
